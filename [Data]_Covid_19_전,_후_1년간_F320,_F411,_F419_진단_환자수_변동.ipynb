# **Covid 19 전, 후 1년간 F320, F411, F419 진단 환자수 변동**

```python
import pandas as pd
import numpy as np
```

## **F320 Data**

```python
f32 = pd.read_excel("/content/f32_원본.xls")
f32.head(5)
```

```python
f32 = pd.read_excel("/content/f32_원본.xls")
f32 = f32.iloc[1:, :]
f32 = f32.rename(columns= {"진료년월" : "기관소재지", "Unnamed: 0" : "코드"})
f32 = f32.reset_index()
f32 = f32.drop("index", axis=1)

f32.head(5)
```

```python
f32_pre = f32.iloc[:, :17]
f32_pre.drop('2018년 11월', axis=1)

f32_pre = f32_pre.transpose()
f32_pre = f32_pre.iloc[1:, :]
f32_pre = f32_pre.rename(columns=f32_pre.iloc[0]).iloc[1:, :]
f32_pre.head(5)
```

```python
locals = f32_pre.columns
locals
```

```python
f32_post = f32.iloc[:, 17:]
f32_post.set_index(locals, inplace=True)
f32_post = f32_post.transpose()
f32_post.head(5)
```

```python
import numpy as np 

for col in locals:
    f32_pre[col] = f32_pre[col].str.replace(",", "")
    f32_pre = f32_pre.astype({col : np.int64})
    f32_post[col] = f32_post[col].str.replace(",", "")
    f32_post = f32_post.astype({col : np.int64})
```

```python
f32_pre_int = f32_pre
f32_post_int = f32_post

f32_pre_int
f32_post_int
```

```python
for col in locals:
  f32_pre_int[col] = f32_pre[col].sum()
  f32_post_int[col] = f32_post[col].sum()
```

```python
# f32_pre_sum 최종

f32_pre_int = f32_pre_int.iloc[:1, :]
f32_pre_int = f32_pre_int.rename(index={'F32' : 'F32_Pre'})
# f32_pre_int
```

```python
# f32_post_sum 최종

f32_post_int = f32_post_int.iloc[:1, :]
# f32_post_int
f32_post_int = f32_post_int.rename(index={'F32' : 'F32_Post'})
```

```python
f32_prevspost = pd.concat([f32_pre_int, f32_post_int])
f32_prevspost = f32_prevspost.transpose()
# f32_prevspost
```

```python
f32_prevspost_total = f32_prevspost.iloc[:1, :]
f32_prevspost_total
f32_prevspost = f32_prevspost.iloc[1:, :]
```

```python
f32_prevspost_total
f32_prevspost
```

```python
with pd.ExcelWriter('f32_prevspost.xlsx') as writer:
  f32_prevspost.to_excel(writer, sheet_name='f32_prevspost')
with pd.ExcelWriter('f32_prevspost_total.xlsx') as writer:
  f32_prevspost_total.to_excel(writer, sheet_name='f32_prevspost_total')
```
